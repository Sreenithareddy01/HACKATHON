```python
df_italian = merged_df[merged_df['cuisines'].str.contains('Italian', case=False, na=False)]
print(df_italian)
```

```python
bangalore_coords = [12.9716, 77.5946]
restaurant_map = folium.Map(location=bangalore_coords, zoom_start=12)
```

```python
df_italian_clean = df_italian.dropna(subset=['Latitude', 'Longitude'])
```

```python
marker_cluster = MarkerCluster().add_to(restaurant_map)
```

```python
for idx, row in df_italian_clean.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"Restaurant {idx}: {row['rest_type']}, Cuisines: {row['cuisines']}", 
        icon=folium.Icon(color='purple', icon='cloud') 
    ).add_to(marker_cluster)
```

```python
for idx, row in df_italian_clean.iterrows():
    city_name = row['listed_incity']
    cuisines = row['cuisines']
    popup_text = f"<b>Restaurant ID:</b> {idx}<br><b>City:</b> {city_name}<br><b>Cuisines:</b> {cuisines}"
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=popup_text,
        icon=folium.Icon(color='purple', icon='cloud')
    ).add_to(marker_cluster)
```

```python
restaurant_map.save("italian_restaurants_details_map.html")
```
