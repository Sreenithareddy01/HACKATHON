```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("zomato_data.csv")
```

```python
df.shape
```

```python
df.info()
```

```python
#Replace '-' values with NaN
df['rate']=df['rate'].replace('-',np.nan)
df
```

```python
df.info()
```

```python
df['rate']
```

```python
# Remove /5 and retain only numeric value.
df['rate']=df['rate'].str.replace('/5','',regex=False)
df
```

```python
df['rate']
```

```python
#Convert to float using pd.to_numeric()
df['rate']=pd.to_numeric(df['rate'],errors='coerce')
df
```

```python
df['rate']
```

```python
#Fill the missing values using median rating.
median_rating=df['rate'].median()
df['rate']=df['rate'].fillna(median_rating)
```

```python
df['rate']
```

```python
df.info()
```

```python
df.to_csv('zomato_data_cleaned.csv',index=False)
df.info()
```

```python
#Step 2: Cost Column (approx_costfor_two_people)
#Remove commas from numeric strings ('1,000' → '1000')
df=pd.read_csv('zomato_data_cleaned.csv')
df['approx_costfor_two_people']=df['approx_costfor_two_people'].astype(str).str.replace(',','')
```

```python
df['approx_costfor_two_people']
```

```python
df.describe()
```

```python
df.info()
```

```python
#convert to numeric
df['approx_costfor_two_people']=pd.to_numeric(df['approx_costfor_two_people'],errors='coerce')
```

```python
df['approx_costfor_two_people']
```

```python
#Fill missing values using the median cost.
df['approx_costfor_two_people']=df['approx_costfor_two_people'].fillna(df['approx_costfor_two_people'].median())
```

```python
df['approx_costfor_two_people']
```

```python
df.info()
```

```python
#Step 3: Categorical Columns
#dish_liked → Replace NaN with "Not Available"
df['dish_liked']=df['dish_liked'].fillna("Not Avaliable")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#cuisines → Replace NaN with "Other"
df['cuisines']=df['cuisines'].fillna("Other")
```

```python
df['cuisines']
```

```python
#rest_type → Replace NaN with "Unknown"
df['rest_type']=df['rest_type'].fillna("Unknown")
```

```python
df['rest_type']
```

```python
df.info()
```

```python
#Step 4: Votes Column
#Fill missing values in votes with median
df['votes']=df['votes'].fillna(df['votes'].median())
```

```python
df['votes']
```

```python
df.info()
```

```python
#Step 5: Binary Encoding
#Convert the following binary fields:
#Column       Mapping
#online_order Yes → 1, No → 0
#book_table   Yes → 1, No → 0
df['online_order']=df['online_order'].map({'Yes':1,'No':0})
df['book_table']=df['book_table'].map({'Yes':1,'No':0})
```

```python
#Step 6: Data Type Conversion
#Ensure the following conversions:
#rate → float
#votes → integer
#approx_cost for_two_people → integer
df['rate']=df['rate'].astype(float)
df['votes']=df['votes'].astype(int)
```

```python
df['approx_costfor_two_people']=df['approx_costfor_two_people'].fillna(df['approx_costfor_two_people'].median())
df['approx_costfor_two_people']=df['approx_costfor_two_people'].astype(int)
```
